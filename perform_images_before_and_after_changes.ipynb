```python
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator
```

before change the images to mode RGB

```python
PATH_CONTROL = '/content/gdrive/MyDrive/mini-project/bacteria_data/data/data_split/validation/control'
PATH_ACID = '/content/gdrive/MyDrive/mini-project/bacteria_data/data/data_split/validation/acid'
```

```python
img_width, img_height = 250, 250 #images size

BATCH_SIZE = 16
control = ImageDataGenerator(rescale=1./255)
acid = ImageDataGenerator(rescale=1./255)


control_data_gen = control.flow_from_directory(batch_size=BATCH_SIZE,
                                                           directory=PATH_CONTROL,
                                                           shuffle=True,
                                                           target_size=(img_width,img_height),
                                                           class_mode='categorical')
acid_data_gen = acid.flow_from_directory(batch_size=BATCH_SIZE,
                                                             directory=PATH_ACID,
                                                             shuffle=False,
                                                             target_size=(img_width,img_height),
                                                             class_mode='categorical')
```

control images:

```python
sample_control_images, _ = next(control_data_gen)
# This function will plot images in the form of a grid with 1 row and 5 columns where images are placed in each column.
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 5, figsize=(20,20))
    axes = axes.flatten()
    for img, ax in zip(images_arr, axes):
        ax.imshow(img)
    plt.tight_layout()
    plt.show()
plotImages(sample_control_images[:5])  # Plot images 0-4
```

acid images:

```python
sample_acid_images, _ = next(acid_data_gen)
# This function will plot images in the form of a grid with 1 row and 5 columns where images are placed in each column.
def plotImages(images_arr):
    fig, axes = plt.subplots(1, 5, figsize=(20,20))
    axes = axes.flatten()
    for img, ax in zip(images_arr, axes):
        ax.imshow(img)
    plt.tight_layout()
    plt.show()
plotImages(sample_acid_images[:5])  # Plot images 0-4
```
