When we started working with photos of bacteria, we ran into a problem. it didn't read the pictures correctly. we checked their mode and got mode=I. so we had to change all photos to RGB format

```python
from google.colab import drive
drive.mount('/content/gdrive')
from PIL import Image
import os
import numpy as np
from tensorflow import keras
from tensorflow.keras.preprocessing.image import img_to_array, load_img
from sklearn.metrics import classification_report
from PIL import Image
```

```python
# paths to original folders control and acidic
# folder paths control acid changed to RGB

PATH_CONTROL = "/content/gdrive/MyDrive/mini-project/bacteria_data/data/control"
PATH_ACID = "/content/gdrive/MyDrive/mini-project/bacteria_data/data/acid"
PATH_control_RGB = "/content/gdrive/MyDrive/mini-project/bacteria_data/data/control_RGB"
PATH_acidic_RGB = "/content/gdrive/MyDrive/mini-project/bacteria_data/data/acidic_RGB"
```

```python
# a function that converts a photo to RGB format
def convert(img, target_type_min, target_type_max, target_type):
    imin = img.min()
    imax = img.max()
    a = (target_type_max - target_type_min) / (imax - imin)
    b = target_type_max - a * imax
    new_img = (a * img + b).astype(target_type)
    return new_img
```

control

```python
if not os.path.exists(PATH_control_RGB):    # checks if the folder exists, if not it creates it.
    os.makedirs(PATH_control_RGB)

# Get a list of all files in the "CONTROL" folder
file_list = os.listdir(PATH_CONTROL)
counter=0     # counter to check if the program is up and running

# Loop through each file
for file_name in file_list:
    if file_name.endswith('.png'):
        print(counter)
        counter+=1
        file_path = os.path.join(PATH_CONTROL, file_name)

        img = Image.open(file_path)
        img_array = np.asarray(img)
        img_array = convert(img_array, 0, 255, np.uint8)    # send the img to the function above
        img = Image.fromarray(img_array)
        output_file = os.path.join(PATH_control_RGB, os.path.splitext(file_name)[0] + '.jpg')

        img.save(output_file)
```

```python
if not os.path.exists(PATH_acidic_RGB):     # checks if the folder exists, if not it creates it.
    os.makedirs(PATH_acidic_RGB)

# Get a list of all files in the "ACID" folder
file_list = os.listdir(PATH_ACID)
counter=0     # counter to check if the program is up and running

# Loop through each file
for file_name in file_list:
    if file_name.endswith('.png'):
        print(counter)
        counter+=1
        file_path = os.path.join(PATH_CONTROL, file_name)

        img = Image.open(file_path)
        img_array = np.asarray(img)
        img_array = convert(img_array, 0, 255, np.uint8)    # send the img to the function above
        img = Image.fromarray(img_array)
        output_file = os.path.join(PATH_control_RGB, os.path.splitext(file_name)[0] + '.jpg')

        img.save(output_file)
```
